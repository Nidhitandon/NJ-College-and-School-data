```python
# Data from NCES and IPEDS


import requests
from datetime import datetime
import seaborn as sns
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt


import warnings
```

```python
df = pd.read_excel('Comcollege.xlsx')
df.tail()
```

```python
df.info()
```

```python
for col in df.columns[1:]:
    print(col)
    print('Skew :',round(df[col].skew(),2))
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    df[col].hist(bins=10, grid=False)
    plt.ylabel('count')
    plt.subplot(1,2,2)
    sns.boxplot(x=df[col])
    plt.show()
```

```python
#plotting individual lines with custom colors, styles, and widths
plt.plot(df['Year'], df['total_students'], label='Total', color='red', linewidth= 2)

plt.plot(df['Year'], df['below_twenty'], label='Below 20', color='grey')
plt.plot(df['Year'], df['twenty_to_twentyfour'], label='20-24', color='steelblue', linestyle='dotted')
plt.plot(df['Year'], df['over_twentyfour'], label='Over 24', color='purple', linestyle='dashed')

#add legend
plt.legend()

#add axis labels and a title
plt.ylabel('Students', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.title('College Enrollment in NJ, 2009-2019', fontsize=16)

#display plot
plt.show()
```

```python

plt.plot(df['Year'], df['total_students'],label='Total College', color='steelblue', linestyle='dotted', linewidth= 4)
plt.plot(df['Year'], df['CC-Enroll'], label='Community College Enrollment', color='purple', linestyle='dashed')


#add legend
plt.legend()

#add axis labels and a title
plt.ylabel('in %', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.title('Total College vs. Community College Enrollment Rates in NJ, 2009-2020', fontsize=16)

#display plot
plt.show()
```

```python

plt.figure(figsize=(10,5))
sns.boxplot(y=df["Percent_persistence_rate"],
            x=df["Year"])
plt.grid(True)
plt.title("% Student Persistence rate by year in NJ",color='black')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(y=df["percent_retention_rate"],
            x=df["Year"])
plt.grid(True)
plt.title("% Student Retention rate by year in NJ",color='black')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(y=df["below_twenty"],
            x=df["Year"])
plt.grid(True)
plt.title("Total Enrollment for Ages 20 and below in NJ",color='black')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(y=df["twenty_to_twentyfour"],
            x=df["Year"])
plt.grid(True)
plt.title("Total Enrollment for Ages 20-24 in NJ",color='black')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(y=df["over_twentyfour"],
            x=df["Year"])
plt.grid(True)
plt.title("Total Enrollment for Ages 24 and over in NJ",color='black')
plt.show()
```
